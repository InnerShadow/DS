```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# fetch dataset 
individual_household_electric_power_consumption = fetch_ucirepo(id=235) 
```

```python
# data (as pandas dataframes) 
X = individual_household_electric_power_consumption.data.features 
y = individual_household_electric_power_consumption.data.targets 
  
# metadata 
#print(individual_household_electric_power_consumption.metadata) 
  
# variable information 
#print(individual_household_electric_power_consumption.variables)
```

```python
x_time = pd.to_datetime(X['Date'] + ' ' + X['Time'], yearfirst = True, dayfirst = False)
```

```python
Global_active_power = pd.to_numeric(X['Global_active_power'], errors = 'coerce')
Sub_metering_1 = pd.to_numeric(X['Sub_metering_1'], errors = 'coerce')
Sub_metering_2 = pd.to_numeric(X['Sub_metering_2'], errors = 'coerce')
Sub_metering_3 = pd.to_numeric(X['Sub_metering_3'], errors = 'coerce')
Y = Global_active_power * 1000 / 60 - Sub_metering_1 - Sub_metering_2 - Sub_metering_3

x_time = x_time.dropna()
Y = Y.loc[x_time.index]


Y = Y.dropna()
x_time = x_time.loc[Y.index]

x_time = x_time[:5000]
Y = Y[:5000]

print(x_time)
print(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_time, Y, test_size=0.2)
```

```python
import numpy as np

print(np.array(X_train).reshape(-1, 1))
print(np.array(y_train).reshape(-1, 1))

model = SVR(kernel='linear', C=1)
model.fit(np.array(X_train).reshape(-1, 1), np.array(y_train))
```

```python
print(model)
```

```python
y_pred = model.predict(np.array(X_test).reshape(-1, 1))

print(np.array(X_test).reshape(-1, 1))

print("\n\n")

print(y_pred)

# Метрики качества
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
```
